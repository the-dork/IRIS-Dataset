```python
# IQR
Q1 = np.percentile(df['sepal_width'], 25,
                interpolation = 'midpoint')
  
Q3 = np.percentile(df['sepal_width'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1
print("Old Shape: ", df.shape)
max =   Q3+1.5*IQR
min =   Q1-1.5*IQR

df['sepal_width'] = np.where(df['sepal_width'] > max, max , df['sepal_width'])
df['sepal_width'] = np.where(df['sepal_width'] < min, min, df['sepal_width'])
print("New Shape: ", df.shape)
  
df.boxplot(column = 'sepal_width')
```
